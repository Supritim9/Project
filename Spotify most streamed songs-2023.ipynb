# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as mp
import numpy as np
import seaborn as sns
```

# Loading Dataset

```python
df = pd.read_csv("spotify-2023.csv",encoding_errors= 'replace')
df.head()
```

# Data Cleaning and Processing

```python
df["in_shazam_charts"] = df["in_shazam_charts"].str.replace(",","").astype(float)
```

```python
df["in_shazam_charts"].mean()
df["in_shazam_charts"] = df["in_shazam_charts"].fillna(60)
```

```python
df["key"] = df["key"].fillna(method = "ffill")
```

```python
df["streams"]= df["streams"].replace("BPM110KeyAModeMajorDanceability53Valence75Energy69Acousticness7Instrumentalness0Liveness17Speechiness3",np.nan)
```

```python
df["streams"] = df["streams"].astype("Int64")
df["streams"].mean()
df["streams"]= df["streams"].fillna(514137425)
```

```python
df["released_date"] = df["released_day"].astype(str)+"/"+ df["released_month"].astype(str)+"/"+ df["released_year"].astype(str)
df["released_date"] = pd.to_datetime(df["released_date"])
df["month_name"] = df["released_date"].dt.strftime("%b")
```

```python
df["streams(Million)"] = (df["streams"] //1000000)
df.head(2)
```

# Data Visualization

# Q1. What are the top 10 most streamed songs on Spotify in 2023?

```python
df = df.sort_values(by = "streams", ascending = False)
top10_songs = df.head(10)
mp.figure(figsize = (6,3))
ax = sns.barplot(data = top10_songs, y = "track_name", x= "streams(Million)", color = "green")
ax.bar_label(ax.containers[0])
mp.title("Top 10 most streamed songs")
mp.show()
```

# Insights: Among Top 10 most Streamed song Blinding Lights is the highest streaming song in 2023 and then Shape of You, Somone You Loved, Dance Monkey, Sunflower, One Dance, Stay, Believer, Closer, and Starboy.

# Q2. Top 10 artist based on streams.

```python
gb = df.groupby("artist(s)_name").agg({"streams(Million)":"max"})
a = gb.sort_values(by = "streams(Million)", ascending = False)
top_10artist = a.head(10)
top_10artist
mp.figure(figsize = (6,3))
ax = sns.barplot(data = top_10artist, y = "artist(s)_name", x= "streams(Million)", color = 'green')
ax.bar_label(ax.containers[0])
mp.title("Top 10 artist based on streams")
mp.show()
```

# Insights: Top 10 artist based on strems are The Weeknd, Ed Sheeran, Lews Capaldi, Tones and I, Post Malon & Swae Lee, Drake, Justin Bieber, Imagine Dragons, The Chainsmokers and The weeknd & Daft Punk.

# Q3. Top 10 artist or group has the most songs in most streamed songs in 2023?

```python
gb = df.groupby("artist(s)_name").agg({"track_name":"count"})
gb
a = gb.sort_values(by = "track_name", ascending = False)
top10_songs = a.head(10)
mp.figure(figsize = (6,3))
ax = sns.barplot(data = top10_songs, y = "artist(s)_name", x= "track_name", color = "green")
ax.bar_label(ax.containers[0])
mp.xlabel("Count of songs")
mp.title("Top 10 artist based on count of songs")
mp.show()
```

# Insights: Taylor Swift has maximum no of songs in most streamed songs and then The weeknd and others.

# Q4. Which artist or group has the most songs in the top 50 most streamed songs in 2023?

```python
top_50 = df.head(50)
gb = top_50.groupby("artist(s)_name").agg({"streams":"sum", "track_name":"count"})
gb.sort_values(by = "track_name", ascending = False)
gb.idxmax()
```

# Insights: Ed Sheeran has the maximum no (4) of songs in the top 50 most streamed songs in 2023.

# Q5. Majorly used key in most streamed songs in 2023?

```python
mp.figure(figsize=(6, 3))
ax = sns.countplot(x="key", data=df, color = "green")
ax.bar_label(ax.containers[0])
mp.ylabel("count of songs")
mp.title("Count of Songs per Key")
mp.show()
```

# Insights: Majorly used key in most streamed songs is C#.

# Q6. Which musical modes appear more frequently among the top streamed songs in 2023?

```python
gb= df.groupby("mode").agg({"track_name":"count"})
gb
mp.figure(figsize = (3,4))
mp.pie(gb["track_name"], labels = gb.index, autopct = "%1.2f", shadow = True, colors = ["green","orange"])
mp.title("Musical modes distribution")
mp.show()
```

# Insights: Major mode(550) is more frequently used among the top streamed songs in 2023.

# Q7. What is the average tempo (BPM) of the most streamed songs in 2023? Are there any outliers?

```python
df["bpm"].mean() #mean = 122.54
mp.figure(figsize= (6,3))
sns.set_style(style = "whitegrid")
ax = sns.boxplot(data = df, x = "bpm", color = "green")
mp.title("BPM distribution of most streamed songs")
mp.show()
```

# Insights: The average beats per minute(BPM) of the most streamed songs is 122.54 and ther is 3 outliers (202, 204, 206).

# Q10. Explore patterns in audio features to understand trends and preferences in popular songs.

```python
df = df.sort_values(by = "streams", ascending = False)
top_10 = df.head(10)
```

```python
mp.figure(figsize = (6,3))
mp.plot(top_10["track_name"], top_10["energy_%"], label = "energy %")
mp.plot(top_10["track_name"], top_10["valence_%"], label = "valence_%")
mp.plot(top_10["track_name"], top_10["liveness_%"], label = "liveness_%")
mp.plot(top_10["track_name"], top_10["instrumentalness_%"], label = "instrumentalness_%")
mp.plot(top_10["track_name"], top_10["danceability_%"], label = "danceability_%")
mp.plot(top_10["track_name"], top_10["acousticness_%"], label = "acousticness_%")
mp.plot(top_10["track_name"], top_10["speechiness_%"], label = "speechiness_%")
mp.legend(bbox_to_anchor = (0.2,0,1.2,1))
mp.xticks(rotation = 90)
mp.xlabel("track_name")
mp.ylabel("Audio features")
mp.title("Audio features of top 10 streamed songs")
mp.show()
```

# Insights: In the top 10 songs Energy is alawyas above 40,Valence is also high that means songs are positive, Liveness is below 40, instrumentalness is 0, danceability is more than 50 that means all songs are danceable, acousticness is high in some songs and low in some songs and lastly speechiness is lower than 10. So top 10 songs are more likely to be more energatic, more danceable and have positive content.
